```python
import numpy as np
import time

python_list = list(range(1000000))
numpy_array = np.arange(1000000)

start = time.time()
sum(python_list)
print("Python list:", time.time() - start)

start = time.time()
numpy_array.sum()
print("NumPy array:", time.time() - start)
```
